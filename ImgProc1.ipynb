The goal is to take pictures below and use image processing to output results: mainly distinguish between two types of cells: vegetative and heterocyst.



These guys organize in a pattern that forms naturally in the simplest organisms - cyanobacteria. We know how to add DNA to bacteria, so can we engineer a change in the pattern, and build our own?

```python
import sklearn
import skimage
from skimage import io
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img1 = skimage.io.imread("C:/users/wsutt_000/desktop/py/ipy4.27/subfolder/IMG_20140504_161239_584.jpg")
img2 = skimage.io.imread("subfolder/IMG_20140504_161309_413.jpg")

f, (plt1,plt2) = plt.subplots(nrows=1,ncols=2, figsize = (14,14))

# plt.autoscale(False)
# plt.subplots.figsize = (15,12)
plt1.imshow(img1)
plt2.imshow(img2)

f.show()
```
